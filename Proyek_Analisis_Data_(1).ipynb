### Pertanyaan 1: Apa metode pembayaran yang paling sering digunakan?

```python
first_payments = payments_df[payments_df['payment_sequential'] == 1]
result = first_payments.groupby('payment_type')['order_id'].count()
total = result.sum()
percentages = (result / total) * 100

# Gabungkan kategori < 5% menjadi "Other"
threshold = 5
result_filtered = result[percentages >= threshold]
other_total = result[percentages < threshold].sum()
if other_total > 0:
    result_filtered['Other'] = other_total
data_chart = result_filtered.index.tolist()

palette_color = sn.color_palette('viridis', len(data_chart))
plt.figure(figsize=(7, 7))
plt.pie(result_filtered, labels=data_chart, colors=palette_color, autopct='%.0f%%')

plt.title('Distribusi Metode Pembayaran (Hanya Pembayaran Pertama)', fontsize=14)
plt.show()
```

## RFM Analysis(Analisis lanjutan)

Dataset tersebut cocok untuk menggunakan analisa RFM(Recency, Frequency dan Moetary) karena memiliki data yang lengkap dan berbasis transaksi. RFM dapat digunakan untuk menemukan pelanggan yang bernilai tinggi. Hasil analisanya dapat digunakan untuk strategi pemasaran dan retensi

```python
rfm_data = orders_df.merge(customers_df, on="customer_id", how="left")
rfm_data = rfm_data.merge(items_df, on="order_id", how="left")
```

```python
rfm_data["order_purchase_timestamp"] = pd.to_datetime(rfm_data["order_purchase_timestamp"])

reference_date = rfm_data["order_purchase_timestamp"].max()

rfm = rfm_data.groupby("customer_unique_id").agg({
    "order_purchase_timestamp": lambda x: (reference_date - x.max()).days,
    "order_id": "count",
    "price": "sum"
}).reset_index()

rfm.columns = ["customer_unique_id", "Recency", "Frequency", "Monetary"]
```

```python
rfm["Recency"] = pd.to_numeric(rfm["Recency"], errors="coerce")
rfm["Frequency"] = pd.to_numeric(rfm["Frequency"], errors="coerce")
rfm["Monetary"] = pd.to_numeric(rfm["Monetary"], errors="coerce")

rfm["Recency"] = rfm["Recency"].fillna(rfm["Recency"].median())
rfm["Frequency"] = rfm["Frequency"].fillna(rfm["Frequency"].median())
rfm["Monetary"] = rfm["Monetary"].fillna(rfm["Monetary"].median())

for col in ["Recency", "Frequency", "Monetary"]:
    if len(rfm[col].unique()) < 4:
        rfm[col] += np.random.uniform(0, 1, size=len(rfm))

rfm["Recency_Rank"] = rfm["Recency"].rank(method="dense") + np.random.uniform(0, 0.01, size=len(rfm))
rfm["Frequency_Rank"] = rfm["Frequency"].rank(method="dense") + np.random.uniform(0, 0.01, size=len(rfm))
rfm["Monetary_Rank"] = rfm["Monetary"].rank(method="dense") + np.random.uniform(0, 0.01, size=len(rfm))

rfm["R_Score"] = pd.qcut(rfm["Recency_Rank"], q=4, labels=[4, 3, 2, 1], duplicates="drop")
rfm["F_Score"] = pd.qcut(rfm["Frequency_Rank"], q=4, labels=[1, 2, 3, 4], duplicates="drop")
rfm["M_Score"] = pd.qcut(rfm["Monetary_Rank"], q=4, labels=[1, 2, 3, 4], duplicates="drop")

rfm["RFM_Score"] = rfm["R_Score"].astype(str) + rfm["F_Score"].astype(str) + rfm["M_Score"].astype(str)

rfm.head()
```

```python
def segment_customer(rfm_score):
    if rfm_score in ["444", "443", "434", "433"]:
        return "Best Customers"
    elif rfm_score in ["344", "334", "333"]:
        return "Loyal Customers"
    elif rfm_score in ["144", "143", "134", "133"]:
        return "New Customers"
    elif rfm_score in ["244", "233", "223"]:
        return "Potential Loyalists"
    elif rfm_score in ["411", "311", "211"]:
        return "Churned Customers"

rfm["Customer_Segment"] = rfm["RFM_Score"].apply(segment_customer)
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=rfm, x="Customer_Segment", order=rfm["Customer_Segment"].value_counts().index, hue="Customer_Segment", palette="viridis", dodge=False)
plt.xticks(rotation=45)
plt.title("Distribusi Segmen Pelanggan")
plt.xlabel("Segmen")
plt.ylabel("Jumlah Pelanggan")
plt.legend([], [], frameon=False)
plt.show()
```
